```python
def compute_mean(numbers):
    if not numbers:
        return 0
    mean_val = sum(numbers) / len(numbers)
    print(mean_val)

compute_mean([2, 4, 6, 8])
```

This code defines a function that calculates the mean of a list by summing all elements and dividing by the total number of elements.

The mean can be expressed in the following ways:

1. 
$$\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$$

2. 
$$\bar{x} = \frac{x_1 + x_2 + \dots + x_n}{n}$$
